```python
# TODO 06 - which Segment has the highest profit in 2018

df2018 = df[pd.DatetimeIndex(df['Order Date']).year == 2018]
dfTopSeg2018 = df2018.groupby('Segment')['Profit'].agg('sum')
indexNames = list(dfTopSeg2018.sort_values(ascending=False).index.values)

print("The segment which has highest profit in 2018 is '",indexNames[0],"'.")
dfTopSeg2018.sort_values(ascending=False)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

dfFix19 = df[(df['Order Date'] >= '2019-4-15') & (df['Order Date'] <= '2020-12-31')]
lowSalesState19 = dfFix19.groupby('State')['Sales'].agg('sum')
low5SalesState19 = lowSalesState19.sort_values().head(5)

indexLowSalesState = list(low5SalesState19.index.values)

print("Top 5 lowest total sales states are",str(indexLowSalesState),"respectively.")
low5SalesState19
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df2019 = df[df["Order Date"].dt.year == 2019]

dfWCen2019 = df2019.query('Region == ["West", "Central"]')
totalSalesWCen2019 = dfWCen2019['Sales'].agg('sum')

totalSales2019 = df2019['Sales'].agg('sum')

resultWCen = (totalSalesWCen2019/totalSales2019)*100

print("Ratio of total sales in West+Central is %.2f"%resultWCen,"%.")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

df19to20 = df[df['Order Date'].dt.year.isin([2019, 2020])]

#Top 10 popular products
dfTopPop19to20 = df19to20.groupby('Product Name')['Sales'].count().sort_values(ascending=False).reset_index().head(10)

#Top 10 total sales
dfTopSales19to20 = df19to20.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).reset_index().head(10)

#Joining DataFrame
df_result19to20 = dfTopPop19to20.join(dfTopSales19to20,lsuffix='_Popular',rsuffix='_Total')
df_result19to20
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# Trend of total Sales growth of each segment from 2017 - 2020

#Count year
df['Year'] = df["Order Date"].dt.year
count_dfYear = df['Year'].value_counts()

#Chart will be plot in the time frame of 2017 - 2020
#Plotting the trend of sales of each segment

df_Pivot = df.pivot_table(index='Year',columns='Segment',values=['Sales'], aggfunc=sum)

import matplotlib.pyplot as plt

df_Pivot.plot()
plt.show()
```
